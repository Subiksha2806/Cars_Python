# Detecting outliers

```python
IQR=q3-q1
IQR
```

```python
lower_limit=q1-(1.5*IQR)
upper_limit=q3+(1.5*IQR)
```

```python
lower_limit
```

```python
upper_limit
```

```python
outliers=cars[(cars["Year"]<lower_limit)|(cars["Year"]>upper_limit)]
outliers
```

```python
newdf=cars[(cars["Year"]>lower_limit)&(cars["Year"]<upper_limit)]
newdf #7195 rows × 14 columns
```
